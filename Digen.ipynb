## **Load the Diverse Generative ML Benchmarkkr(DIGEN)**

```python
#Import a package and initialize the benchmark
from digen import Benchmark
import pandas as pd
from m5gp import m5gpRegressor
from m5gp import m5gpClassifier
import logging

# Desactivar los mensajes de registro
logging.getLogger('cuml').setLevel(logging.ERROR)

benchmark = Benchmark()
```

```python
est = m5gpClassifier()


def parameters(trial):
    params = {
        'generations': 30, # 30
        'Individuals': 256, # 256
        'GenesIndividuals': trial.suggest_categorical('value', [16, 32, 64, 128]), # 32,64,128
        'mutationProb': 0.1,
        'mutationDeleteRateProb': 0.01,
        'sizeTournament': 0.15,
        'evaluationMethod': 0,
        'maxRandomConstant': 999,
        'genOperatorProb': 0.50,
        'genVariableProb': 0.39,
        'genConstantProb': 0.1,
        'genNoopProb': 0.01,
        'verbose': 1,
        'useOpIF': 0,
        'crossVal': True,
        'k': 3,
        'scorer': 0
    }

    return params
```

```python
results = benchmark.optimize(est = est, parameter_scopes = parameters, datasets=["digen10_8322"])
```
